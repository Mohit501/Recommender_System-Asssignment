```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('/kaggle/input/movielens100k/movies.csv')
movies
```

```python
len(np.unique(movies.movieId))
```

```python
ratings = pd.read_csv('/kaggle/input/movielens100k/ratings.csv')
```

```python
ratings
```

```python
user_ratings = {}

for i in range(ratings.shape[0]):
    userid = ratings.iloc[i].userId
    movieId = ratings.iloc[i].movieId
    rating = ratings.iloc[i].rating
    
    if userid in user_ratings.keys():
        user_ratings[userid][movies[movies['movieId'] == movieId].index[0]] = rating
    else:
        user_ratings[userid] = [0 for _ in range(9125)]
```

```python
main_df = pd.DataFrame(user_ratings).T
```

```python
main_df
```

```python
similarity_array = cosine_similarity(np.array(main_df))
```

```python
def get_movies_recommend(user_id, similarity_array):
    
    temp_list = list(np.argsort(similarity_array[user_id])[-11:][::-1])[1:]
    
    temp = []
    for movie in range(9125):
        sim,s  = 0, 0
        for i in list(np.argsort(similarity_array[user_id])[-11:][::-1])[1:]:
            s += main_df.iloc[i][movie]
            sim += similarity_array[i][user_id]

        temp.append(s/sim)
        
    temp2 = np.argsort(np.array(temp))[-10:][::-1]
    
    for i in temp2:
        print(movies.iloc[i].title)
```

```python
get_movies_recommend(246, similarity_array)
```
